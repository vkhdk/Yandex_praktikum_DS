```python
def na_values (data):
    report = data.isna().sum().to_frame()
    report = report.rename(columns = {0: 'missing_values'})
    report['% of total'] = (report['missing_values'] / data.shape[0]).round(2)
    return report.sort_values(by = 'missing_values', ascending = False)
```

Проверим гипотезу: Временной ряд не является стационарным. 
    
    Нулевая гипотеза (H0): Стационарный ряд не является стационарным.
    Альтернативная гипотеза (H1): Стационарный ряд является стационарным.
    Критический уровень статистической значимости "alpha" - 0.05
    
Для проверки гипотезы воспользуемся дополненным тестом Дики-Фуллера.

```python
alpha = 0.05
values_seria = data_resample.values
result = adfuller(values_seria)
if result[1] < alpha:
    print(f'Отвергаем нулевую гипотезу.\np-value: {result[1]}')
else:
    print(f'Не получилось отвергнуть нулевую гипотезу.\np-value: {result[1]}')
```

У нас нет оснований подтвердить гипотезу о том, что ряд не является стационарным. Считаем, что ряд стационарный.

### 3. Построение моделей.<a id='step3'></a>

#### 3.1 Создание признаков.<a id='step3.1'></a>

Создадим признаки для горизонта прогнозирования в один шаг. В качестве признаков будем использовать три вида признаков:

- "Календарные признаки" - Час в который сделан заказ и день недели.
- "Отстающие значения" - Предыдущие значения с отставанием в "n" шагов.
- "Скользящее среднее" - Усредненное значение за "z" шагов.

Напишем функцию для создания признаков.

```python
def make_features(data, max_lag, rolling_mean_size):
    data['dayofweek'] = data.index.dayofweek
    data['hour'] = data.index.hour    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
```

Создадим копию исходного набора данных и добавим в копию 2 признака "отстающего значения" и скользящее среднее за 24 шага.

```python
data_resample_enriched = data_resample.copy()
make_features(data_resample_enriched,2,24)
```

Оценим получившуюся таблицу.

```python
data_resample_enriched.head(15)
```

Получить признаки для первых значений нельзя: исторических данных по ним нет. Удалим строки содержащие пропущенные значения.

```python
data_resample_enriched = data_resample_enriched.dropna()
```

#### 3.2 Получение обучающего и тестового набора данных.<a id='step3.2'></a>

Заказчик не предоставил тестовый набор данных. Подбор оптимальных гипперпараметров будет выполняться с использованием "GridSearchCV" поэтому нет необходимости делить исходную выборку на обучающую, валидационную и тестовую. Разделим исходную выборку на обучающий и тестовый наборы данных используя функцию "train_test_split".

```python
features_train, features_test, target_train, target_test = train_test_split(
    data_resample_enriched.drop('num_orders', axis=1), 
    data_resample_enriched.num_orders, shuffle=False, test_size=0.1)
```

#### 3.3 Получение оптимальных гиперпараметров.<a id='step3.3'></a>

В исследовании будут сравниваться несколько моделей:
- Случайный лес
- Линейная регрессия
- Градиентный бустинг "LGBMRegressor"
- Простая модель "DummyRegressor"

Создадим список моделей и добавим туда модели которые не требуют подбора гипперпараметров.

```python
models_list = [LinearRegression(),
               DummyRegressor(strategy="median")]
```

Модели "случайный лес" и градиентного бустинга "LGBMRegressor" требуют подбора оптимальных гипперпараметров. У LightGBM есть множество гиперпараметров, в данном проекте будут использоваться только количество деревьев и максимальная глубина. Подберем оптимальные гиперпараметры для моделей используя функцию "GridSearchCV". Согласно требованию заказчика качество предсказания определяется функцией RMSE. RMSE пропорционально MSE и поэтому в качестве функции качества будем использовать MSE. Создадим словари с параметрами для "GridSearchCV".

```python
parametrs_LightGBM = {'n_estimators': [10, 50, 100, 200], 'max_depth': range (2, 5)}
parametrs_forest = { 'n_estimators': [10, 50, 100, 200], 'max_depth': range (2, 5) }
```

Подбор оптимальных гипперпараметров для модели градиентного бустинг "LGBMRegressor". Наилучшую модель добавим в конец списка "models_list".

```python
%%time
model = LGBMRegressor()
tscv = TimeSeriesSplit(n_splits=8)
grid = GridSearchCV(model,parametrs_LightGBM,scoring='neg_mean_squared_error',cv=tscv)
grid.fit(features_train,target_train)
print(f'Лучшие гиперпараметры для модели {model} - {grid.best_params_}')
models_list.append(grid.best_estimator_)
```

Подбор оптимальных гипперпараметров для модели "случайный лес". Наилучшую модель добавим в конец списка "models_list".

```python
%%time
model = RandomForestRegressor(random_state=12)
tscv = TimeSeriesSplit(n_splits=8)
grid = GridSearchCV(model,parametrs_LightGBM,scoring='neg_mean_squared_error',cv=tscv)
grid.fit(features_train,target_train)
print(f'Лучшие гиперпараметры для модели {model} - {grid.best_params_}')
models_list.append(grid.best_estimator_)
```

#### 3.4 Оценка качества предсказаний на тестовом наборе данных.<a id='step3.4'></a>

Оценим получившийся список моделей.

```python
models_list
```

Напишем функцию для расчета RMSE на тестовом наборе данных используя наилучшие модели. Функция выдает значение метрики, длительность обучения модели, график десяти важнейших признаков при предсказании, а так же график сравнения предсказанных значений и фактических. Чтобы более детально сравнить предсказанные и фактические значения выведем данные за три дня (первые 72 часа)

```python
def rmse_calculation(model,features_train,target_train,features_test,target_test):
    start_time = time.time()
    model.fit(features_train, target_train)
    finish_time = time.time()
    predicted = model.predict(features_test)
    mse = mean_squared_error(target_test, predicted)
    
    print(f'--------------------------')
    print(f'Значение RMSE для модели {model} на тестовом наборе данных: {mse ** 0.5}')
    print(f'Время обучения составляет: {finish_time - start_time} секунд\n')
    graf_pred = pd.Series(data=predicted, index=target_test.index)
    graf_target = target_test
    graf_target.head(72).plot(figsize=(15,7),alpha=0.5, label='target')
    graf_pred.head(72).plot(alpha=0.7, label='predicted')
    plt.legend()
    plt.ylabel('Количество заказов.',fontsize=15)
    plt.xlabel('Дата.',fontsize=15)
    plt.title(f'Сравнение значений целевого признака и предсказанных значений.', fontsize=15)
    plt.grid()
    plt.show()
    
    try:
        feature_names = list(features_train)
        importances = model.feature_importances_
        model_importances = pd.Series(importances, index=feature_names)
        model_importances = model_importances.sort_values(ascending=False).head(10)
        model_importances.plot.bar(figsize=(10,5))
        plt.xlabel('Признаки.', fontsize=15)
        plt.ylabel('Важность.',fontsize=15)
        plt.title(f'Топ-10 важных признаков модели.', fontsize=15)
        plt.grid()
        plt.show()
        print(f'--------------------------')
        
    except:
        print(f'--------------------------')
```

```python
for model in models_list:
    rmse_calculation(model,features_train,target_train,features_test,target_test)
```
